```python
import pandas as pd

# Cargar CSV principal
df_final = pd.read_csv("final.csv")

# Cargar CSV que contiene las imágenes
df_imgs = pd.read_csv("editar.csv")[['Player', 'PlayerImg', 'TeamImg']]

# Merge por nombre del jugador
df_final = df_final.merge(
    df_imgs,
    on='Player',
    how='left'
)

# Guardar CSV final con imágenes
df_final.to_csv("final1.csv", index=False)

print("PlayerImg y TeamImg añadidos correctamente a final.csv")
```
