```python
import pandas as pd
df = pd.read_csv('dataset_con_clusters_y_estilos.csv')
```

```python
league_metrics = df.groupby('Comp').agg({
    'PrgDist':'mean',
    'PrgP':'mean',
    'Carries':'mean',
    'xG':'mean',
    'Gls':'mean',
    'Tkl+Int':'mean',
    'FatigueIndex':'mean'
})
```

```python
league_norm = (league_metrics - league_metrics.mean()) / league_metrics.std()
```

```python
def league_insight(row):
    if row['xG'] > 1:
        return 'Liga ofensiva y vertical'
    if row['PrgP'] > 1:
        return 'Liga táctica y de posesión'
    if row['Tkl+Int'] > 1:
        return 'Liga física y defensiva'
    return 'Liga equilibrada'

league_norm['StyleInsight'] = league_norm.apply(league_insight, axis=1)
league_norm[['StyleInsight']]
```

```python
league_norm.to_csv('league_profiles.csv')
```
