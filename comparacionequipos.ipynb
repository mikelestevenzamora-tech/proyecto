```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('league_profiles.csv')
df.head()
```

```python
team_metrics = [
    'PrgDist',        # verticalidad
    'PrgP',           # progresión por pase
    'Carries',        # conducción
    'Att_stats_possession',
    'xG',
    'xAG',
    'Gls',
    'Tkl+Int',
    'Blocks',
    'FatigueIndex'
]
```

```python
print(df.columns.tolist())
```

```python
team_profiles = df.groupby('Squad')[team_metrics].mean()
team_profiles.head()
```

```python
team_profiles_norm = (team_profiles - team_profiles.mean()) / team_profiles.std()
team_profiles_norm.head()
```

```python
team_profiles_norm.sort_values('PrgDist', ascending=False).head(10)
```

```python
team_profiles_norm.sort_values('Tkl+Int', ascending=False).head(10)
```

```python
def team_insights(row):
    insights = []

    # Insights fuertes
    if row['xG'] > 1:
        insights.append('Alta producción ofensiva')
    if row['PrgDist'] > 1:
        insights.append('Juego vertical')
    if row['PrgP'] > 1:
        insights.append('Progresión por pase')
    if row['Carries'] > 1:
        insights.append('Uso intensivo de conducción')
    if row['Tkl+Int'] > 1:
        insights.append('Alta intensidad defensiva')
    if row['FatigueIndex'] > 1:
        insights.append('Alto desgaste físico')
    if len(insights) == 0:
        if row['xG'] > 0:
            insights.append('Perfil ofensivo equilibrado')
        elif row['Tkl+Int'] > 0:
            insights.append('Perfil defensivo equilibrado')
        else:
            insights.append('Perfil táctico balanceado')

    return ', '.join(insights)
```

```python
team_summary = team_profiles_norm.copy()
team_summary['Style_Insight'] = team_profiles_norm.apply(team_insights, axis=1)

team_summary[['Style_Insight']].head(10)
```

```python
team_summary.to_csv('team_profiles_big5.csv')
```
